```python
import cv2
import sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
img  = cv2.imread('./Desktop/KakaoTalk_20220707_222728324.jpg')

if img is None:
    print('image read failed')
    sys.exit()
    
cv2.namedWindow('image')
cv2.imshow('image', img)
    
cv2.waitKey()

cv2.destroyAllWindows()
```

```python
img = cv2.imread('./Desktop/KakaoTalk_20220707_222728324.jpg', cv2.IMREAD_GRAYSCALE)

if img is None:
    print('image read failed')
    sys.exit()
    
    
cv2.namedWindow('image')
cv2.imshow('image', img)

cv2.waitKey()

cv2.destroyAllWindows()
```

```python
img = cv2.imread('./Desktop/KakaoTalk_20220707_222728324.jpg', cv2.IMREAD_COLOR)

if img is None:
    print('image read failed')
    sys.exit()
    
plt.figure(figsize = (20, 20))
plt.subplot(131), plt.imshow(img), plt.axis('off'), plt.title('baseball')
plt.show()
```

```python
import cv2
import sys
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
img = cv2.imread('./fig/PD_MRI.jpeg')

if img is None:
    print('image read failed')
    sys.exit()
    
cv2.namedWindow('image', cv2.WINDOW_AUTOSIZE)    
cv2.imshow('image', img)
    
while True:
    key = cv2.waitKey()
    if key == 27 or key == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
img = cv2.imread('./fig/field.bmp')

if img is None:
    print('image read failed')
    sys.exit()
    
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# plt.imshow(img)
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
