```python
import cv2
import sys
import numpy as np

import glob
import matplotlib.pyplot as plt
```

## 히스토그램 - 흑백

```python
img = cv2.imread('./fig/lenna.bmp', 0)

# calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]]) -> hist
# images: 입력영상 리스트(리스트로 입력)
# channels: 채널리스트, 3 채널의 경우 [0,1,2]
# mask: 마스크 영상입력, 영상전체는 None으로 지정
# histSize: 히스터그램 빈의 크기
# range: 히스토그램의 최솟값과 최댓값
# hist: 계산된 히스토그램, numpy.ndarray 타입
# accumulate: 기존의 히스토그램을 누적할 경우   True

hist = cv2.calcHist([img], [0], None, [256], [0, 256])

plt.plot(hist)
plt.show()

cv2.imshow('image', img)

cv2.waitKey()
cv2.destroyAllWindows()
```

## 히스토그램 - 컬러

```python
img = cv2.imread('./fig/lenna.bmp', 1)

# calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]]) -> hist
# images: 입력영상 리스트(리스트로 입력)
# channels: 채널리스트, 3 채널의 경우 [0,1,2]
# mask: 마스크 영상입력, 영상전체는 None으로 지정
# histSize: 히스터그램 빈의 크기
# range: 히스토그램의 최솟값과 최댓값
# hist: 계산된 히스토그램, numpy.ndarray 타입
# accumulate: 기존의 히스토그램을 누적할 경우   True

hist_b = cv2.calcHist([img], [0], None, [256], [0, 256])
hist_g = cv2.calcHist([img], [1], None, [256], [0, 256])
hist_r = cv2.calcHist([img], [2], None, [256], [0, 256])

plt.plot(hist_b, color = 'b')
plt.plot(hist_g, color = 'g')
plt.plot(hist_r, color = 'r')

plt.show()

cv2.imshow('image', img)

cv2.waitKey()
cv2.destroyAllWindows()
```
