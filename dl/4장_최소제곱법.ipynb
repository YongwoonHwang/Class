```python
import numpy as np 
```

```python
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])
```

```python
mx=np.mean(x)
my=np.mean(y)
```

```python
divisor = sum([(i-mx)**2 for i in x])
```

```python
def top(x, mx, y, my):
  d = 0
  for i in range(len(x)):
    d += (x[i] - mx) * (y[i] -my)
  return d
dividend = top(x, mx, y, my)
```

```python
a = dividend / divisor
b = my - (mx * a)
```

```python
print('x 평균 : ', mx)
print('y 평균 : ', my)
print('\n')
print('분자 : ', dividend)
print('분모 : ', divisor)
print('\n')
print('기울기 : ', a)
print('y 절편 : ', b)
```

```python
fake_a = 3
fake_b = 76
```

```python
def predict(x):
  return fake_a * x + fake_b
```

```python
predict_result = []
for i in range(len(x)):
  predict_result.append(predict(x[i]))
  print('공부 시간=%.f, 실제 점수=%.f, 예측점수=%.f' %(x[i], y[i], predict(x[i])))
```

```python
n = len(x)
def mse(y, y_pred):
  return (1/n) * sum((y-y_pred)**2)
print('평균 제곱 오차: '+ str(mse(y, predict_result)))
```

# 선형 회귀모델

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.show()
```

```python
a = 0 
b = 0
```

```python
lr = 0.03
```

```python
epochs = 2001
```

```python
n = len(x)
```

```python
# 경사 하강법
for i in range(epochs):
  y_pred = a*x + b
  error = y - y_pred

  a_diff = (2/n) * sum(-x * (error))
  b_diff = (2/n) * sum(-(error))

  a = a - lr * a_diff
  b = b - lr * b_diff

  if i % 100 == 0:
    print('epoch=%.f, 기울기=%.04f, 절편=%.04f' % (i, a, b))
    y_pred = a*x+b
    plt.scatter(x,y)
    plt.plot(x, y_pred, 'r')
    plt.show()
```

Tensorflow선형회귀

```python
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = np.array([2,4,6,8])
y = np.array([81, 93, 91, 97])

model = Sequential()
model.add(Dense(1, input_dim=1, activation='linear'))
model.compile(optimizer='sgd', loss='mse')
model.fit (x, y, epochs = 2000)
```

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r')
plt.show()
```

```python
hour=7
prediction = model.predict([hour])
print('%.f시간을 공부할 경우의 예상 점수는 %.02f점입니다.'%(hour,prediction))
```

## Tensorflow 다중회귀

```python
x = np.array([[2,0], [4,4], [6,2], [8,3]])
y = np.array([81, 93, 91, 97])

model = Sequential()

model.add(Dense(1, input_dim=2, activation='linear'))
model.compile(optimizer='sgd', loss = 'mse')
model.fit(x,y, epochs=2000)
```

```python
hour =7
private_class = 4
prediction = model.predict([[hour, private_class]])
print('%.f시간을 공부하고 %.f시간의 과외를 받을 경우, 예상 점수는 %.02f점입니다.'%(hour, private_class, prediction))
```

##Tensorflow 로지스틱회귀

```python
x = np.array([2,4,6,8,10,12,14])
y = np.array([0,0,0,1,1,1,1])
```

```python
# 모델마다 옵티마이즈 함수와 손실함수를 페어로 외워줄 것 ex) 시그모이드 로지스틱회귀 - sgd - binary_crossentropy

model= Sequential()
model.add(Dense(1, input_dim=1, activation='sigmoid'))
model.compile(optimizer='sgd', loss='binary_crossentropy')
model.fit(x,y,epochs=5000)
```

```python
plt.scatter(x,y)
plt.plot(x, model.predict(x), 'r')
plt.show()
```

```python
hour=7
prediction= model.predict([hour])
print('%.f시간을 공부할 경우, 합격 예상 확률은 %.01f%%입니다.'%(hour, prediction*100))
```
